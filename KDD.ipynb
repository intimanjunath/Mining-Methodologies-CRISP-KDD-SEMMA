#2. Preprocessing (Data Cleaning)


```python
print(df.columns)
```

```python
# Strip whitespace and ensure the column exists
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
if 'protocol' in df.columns:
    df['protocol'].fillna(df['protocol'].mode()[0], inplace=True)
else:
    print("'protocol' column is not present in the dataset")
```

```python
for col in df.columns:
    if df[col].dtype == 'object':  # For categorical columns
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:  # For numeric columns
        df[col].fillna(df[col].mean(), inplace=True)
```

#3. Transformation (Feature Engineering)

```python
# Strip whitespace and ensure the column exists
df.columns = df.columns.str.strip()

# Check if the columns exist in the DataFrame
if 'src_ip' in df.columns and 'dst_ip' in df.columns:
    df = df.drop(columns=['src_ip', 'dst_ip'], axis=1)
    print("Columns 'src_ip' and 'dst_ip' dropped successfully.")
else:
    missing_columns = [col for col in ['src_ip', 'dst_ip'] if col not in df.columns]
    print(f"Columns not found in DataFrame: {missing_columns}")

    # Print all columns to help identify potential typos
    print("Available columns:", df.columns)
```

#4. Data Mining (Model Building)

```python
# Print all column names
print(df.columns)
```

```python
# Encoding the 'proto' column
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder instance
le = LabelEncoder()

# Apply Label Encoding to the 'proto' column
df['proto'] = le.fit_transform(df['proto'])

# Check the transformation
print(df['proto'].head())
```

```python
categorical_columns = ['proto', 'service', 'state']  # Add any other categorical columns

for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.model_selection import train_test_split

# Define the feature set (X) and the target (y)
X = df.drop(['label', 'attack_cat'], axis=1)  # Drop the target columns
y = df['label']  # The label column is your target (0: normal, 1: attack)

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training set
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model performance
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")

# ROC-AUC score
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print(f"ROC-AUC Score: {roc_auc:.4f}")

# Plot ROC curve
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
plt.figure()
plt.plot(fpr, tpr, label=f"ROC curve (area = {roc_auc:.2f})")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.show()
```

```python
# Import numpy for array manipulation
import numpy as np

# Feature importance
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")
for i in range(X_train.shape[1]):
    print(f"{i + 1}. Feature {X_train.columns[indices[i]]} ({importances[indices[i]]})")
```
